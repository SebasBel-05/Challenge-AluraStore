```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#**1.Analisis de facturacion**

```python
def ingreso_tienda (dataset):
  return dataset['Precio'].sum()
```

*##Tienda 1*

```python
ingreso_total_tienda1 = ingreso_tienda(tienda)
ingreso_total_tienda1
```

##Tienda 2

```python
ingreso_total_tienda2 = ingreso_tienda(tienda2)
ingreso_total_tienda2
```

##Tienda 3



```python
ingreso_total_tienda3 = ingreso_tienda(tienda3)
ingreso_total_tienda3
```

##Tienda 4

```python
ingreso_total_tienda4 = ingreso_tienda(tienda4)
ingreso_total_tienda4
```

###Datos de las respectivas tiendas

```python
def resultados_datos(series1, series2, series3, series4, nombreColumna):
  series = {'Tienda 1': series1, 'Tienda 2': series2, 'Tienda 3': series3, 'Tienda 4': series4}
  return pd.Series(series, name=nombreColumna)
```

```python
ingresos_totales = resultados_datos(ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4, 'Ingresos totales').astype(int)
ingresos_totales
```

###Grafica

```python
fig, ingresos = plt.subplots(figsize=(10,5))
ingresos.bar(ingresos_totales.index, ingresos_totales.values)
ingresos.set_title("Ingresos totales por tienda", fontsize=16, fontweight='bold')
ingresos.set_xlabel("Tienda")
ingresos.set_ylabel("Ingreso total")
for i, valor in enumerate(ingresos_totales.values):
  ingresos.text(i, valor+1, f'{valor:,.0f}', ha='center', va='bottom', fontsize = 10, color='black' )
  ingresos.ticklabel_format(style='plain', axis='y')
```

#**2.Venta por categoria**

#**4.Productos mas y menos vendidos**

```python
def productos_mas_vendidos(dataset):
  return dataset.groupby('Producto').size().reset_index(name='Cantidad Vendida').sort_values('Cantidad Vendida', ascending=False).head(3)
```

```python
def productos_menos_vendidos(dataset):
  return dataset.groupby('Producto').size().reset_index(name='Cantidad Vendida').sort_values('Cantidad Vendida', ascending=True).head(3)
```

###Tienda 3 (3 productos más vendidos)

###Tienda 3 (3 productos menos vendidos)

###Grafica

```python
fig3, productos= plt.subplots(4,1, figsize=(12,10))
fig3.suptitle('Categoría de producto más y menos vendida por tienda', fontsize = 16, fontweight='bold')

productos[0].barh(productos_menos_vendidos_tienda1['Producto'], productos_menos_vendidos_tienda1['Cantidad Vendida'], color='red', alpha=0.8)
productos[0].barh(productos_mas_vendidos_tienda1['Producto'], productos_mas_vendidos_tienda1['Cantidad Vendida'], color='green', alpha=0.8)
productos[0].set_title('Tienda 1')

productos[1].barh(productos_menos_vendidos_tienda2['Producto'], productos_menos_vendidos_tienda2['Cantidad Vendida'], color='red', alpha=0.8)
productos[1].barh(productos_mas_vendidos_tienda2['Producto'], productos_mas_vendidos_tienda2['Cantidad Vendida'], color='green', alpha=0.8)
productos[1].set_title('Tienda 2')

productos[2].barh(productos_menos_vendidos_tienda3['Producto'], productos_menos_vendidos_tienda3['Cantidad Vendida'], color='red', alpha=0.8)
productos[2].barh(productos_mas_vendidos_tienda3['Producto'], productos_mas_vendidos_tienda3['Cantidad Vendida'], color='green', alpha=0.8)
productos[2].set_title('Tienda 3')

productos[3].barh(productos_menos_vendidos_tienda4['Producto'], productos_menos_vendidos_tienda4['Cantidad Vendida'], color='red', alpha=0.8)
productos[3].barh(productos_mas_vendidos_tienda4['Producto'], productos_mas_vendidos_tienda4['Cantidad Vendida'], color='green', alpha=0.8)
productos[3].set_title('Tienda 4')
productos[3].set_xlabel('Cantidad Vendida')

for ax in productos:
  for bars in ax.containers:
    ax.bar_label(bars,fmt='%.0f',label_type='edge',fontsize=9)


fig3.legend(['Menos vendidos', 'Más vendidos'], loc='upper right')
plt.tight_layout(rect=[0,0,1,0.96]) # Para ajustar la imagen y que no se superpongan elementos
```
